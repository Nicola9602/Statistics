```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro, kurtosis, levene, skewnorm
```

```python
# Normality check with Shapiro-Wilk

def test_normality(data, description):
    stat, p = shapiro(data)
    print(f"Normality Test for {description}:")
    print(f"  Test Statistic: {stat:.4f}, p-value: {p:.4f}")
    if p > 0.05:
        print("  Data appears to be normally distributed.\n")
    else:
        print("  Data does NOT appear to be normally distributed.\n")
```

```python
## Function to calculate kurtosis
def calculate_kurtosis(data, description):
    kurt = kurtosis(data)
    print(f"Kurtosis for {description}: {kurt:.4f}")
    if kurt > 0:
        print("  Distribution has heavier tails (Leptokurtic).\n")
    elif kurt < 0:
        print("  Distribution has lighter tails (Platykurtic).\n")
    else:
        print("  Distribution has normal tails (Mesokurtic).\n")
```

```python
# Function to perform Levene's test for equality of variances
def test_levenes(data1, data2, desc1, desc2):
    stat, p = levene(data1, data2)
    print(f"Levene's Test between {desc1} and {desc2}:")
    print(f"  Test Statistic: {stat:.4f}, p-value: {p:.4f}")
    if p > 0.05:
        print("  Variances appear equal.\n")
    else:
        print("  Variances are significantly different.\n")
```

```python
# Generate distirbutions
# 1. Normal Distribution
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# 2. Positively Skewed Distribution
pos_skewed_data = skewnorm.rvs(a=10, size=1000)

# 3. Negatively Skewed Distribution
neg_skewed_data = skewnorm.rvs(a=-10, size=1000)

# 4. High Kurtosis (Leptokurtic) Distribution
leptokurtic_data = np.concatenate([np.random.normal(0, 0.5, size=900), np.random.normal(0, 5, size=100)])

# 5. Low Kurtosis (Platykurtic) Distribution
platykurtic_data = np.random.uniform(low=-2, high=2, size=1000)
```

```python
# Plot the distributions
plt.figure(figsize=(15, 8))

# Normal Distribution
plt.subplot(2, 3, 1)
plt.hist(normal_data, bins=30, density=True, alpha=0.7, color='blue')
plt.title('Normal Distribution')

# Positively Skewed Distribution
plt.subplot(2, 3, 2)
plt.hist(pos_skewed_data, bins=30, density=True, alpha=0.7, color='green')
plt.title('Positively Skewed Distribution')

# Negatively Skewed Distribution
plt.subplot(2, 3, 3)
plt.hist(neg_skewed_data, bins=30, density=True, alpha=0.7, color='red')
plt.title('Negatively Skewed Distribution')

# High Kurtosis Distribution
plt.subplot(2, 3, 4)
plt.hist(leptokurtic_data, bins=30, density=True, alpha=0.7, color='purple')
plt.title('High Kurtosis (Leptokurtic)')

# Low Kurtosis Distribution
plt.subplot(2, 3, 5)
plt.hist(platykurtic_data, bins=30, density=True, alpha=0.7, color='orange')
plt.title('Low Kurtosis (Platykurtic)')

plt.tight_layout()
plt.show()
```

```python
#Check normality kurtosis THIS IS DONE ON ONLY ONE DISTRIBUTION 
# 1. Normal Distribution
test_normality(normal_data, "Normal Distribution")
calculate_kurtosis(normal_data, "Normal Distribution")

# 2. Positively Skewed Distribution
test_normality(pos_skewed_data, "Positively Skewed Distribution")
calculate_kurtosis(pos_skewed_data, "Positively Skewed Distribution")

# 3. Negatively Skewed Distribution
test_normality(neg_skewed_data, "Negatively Skewed Distribution")
calculate_kurtosis(neg_skewed_data, "Negatively Skewed Distribution")

# 4. High Kurtosis (Leptokurtic) Distribution
test_normality(leptokurtic_data, "High Kurtosis (Leptokurtic) Distribution")
calculate_kurtosis(leptokurtic_data, "High Kurtosis (Leptokurtic) Distribution")

# 5. Low Kurtosis (Platykurtic) Distribution
test_normality(platykurtic_data, "Low Kurtosis (Platykurtic) Distribution")
calculate_kurtosis(platykurtic_data, "Low Kurtosis (Platykurtic) Distribution")

# Perform Levene's test to compare variances THIS IS DONE AS A COMPARISON BETWEEN TWO DISTIRBUTIONS
test_levenes(normal_data, pos_skewed_data, "Normal Distribution", "Positively Skewed Distribution")
test_levenes(normal_data, neg_skewed_data, "Normal Distribution", "Negatively Skewed Distribution")
test_levenes(leptokurtic_data, platykurtic_data, "Leptokurtic Distribution", "Platykurtic Distribution")


```
